```python
import matplotlib.pyplot as plt
import numpy as np

for i in range(50):
    filename = f'../build/results_{i}.csv'
    data = np.loadtxt(filename, delimiter=',', skiprows=2)
    x = data.transpose()[0]
    rho = data.transpose()[1]
    plt.plot(x, rho, 'o--')

plt.grid()
plt.ylim(0, 2)
plt.ylabel('$\\rho$ (kg/m$^3$)')
plt.show()
```
