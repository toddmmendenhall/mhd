```python
import matplotlib.pyplot as plt
import numpy as np

for i in range(8):
    filename = f'../build/test/results_{i}.csv'
    data = np.loadtxt(filename, delimiter=',', skiprows=2)
    x = data.transpose()[0][:-2]
    rho = data.transpose()[8][:-2]
    plt.plot(x, rho, 'o--')

plt.grid()
# plt.ylim(0, 2)
plt.ylabel('$\\rho$ (kg/m$^3$)')
plt.show()
```
