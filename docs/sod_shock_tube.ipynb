```python
import matplotlib.pyplot as plt
import numpy as np

labels = [0, 19.3, 39.0]

for ii, i in enumerate([0, 100, 200]):
    filename = f'../build/results_{i}.csv'
    data = np.loadtxt(filename, delimiter=',', skiprows=2)
    # x = data.transpose()[0]
    rho = data.transpose()[1][200:-206]
    plt.plot(rho, '--', label=f'{labels[ii]} ms')
plt.grid()
plt.legend()
plt.ylim(0, 2)
plt.ylabel('$\\rho$ (kg/m$^3$)')
plt.show()
```
