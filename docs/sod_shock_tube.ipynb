```python
import matplotlib.pyplot as plt
import numpy as np

files = ['../build/test_1/results_0.csv', '../build/test_1/results_99.csv', '../build/test_2/results_99.csv', '../build/test_3/results_99.csv',
         '../build/test_4/results_99.csv', '../build/test/results_99.csv']
labels = ['Initial', '1st order', '2nd order', '3rd order', '2nd order - Van Leer', 'blah']

for i, (file, label) in enumerate(zip(files, labels)):
    if i == 3:
        continue
    data = np.loadtxt(file, delimiter=',', skiprows=2)
    x = data.transpose()[0][:-2]
    y = data.transpose()[1][:-2]
    plt.plot(x, y, label=label)

plt.legend(title='$t=1$ s')
plt.xlabel('x (m)')
plt.ylabel('$\\rho$ (kg/m$^3$)')
plt.show()
```

```python
import matplotlib.animation as animation

fig, ax = plt.subplots()

def update(frame):
    # for each frame, update the data stored on each artist.
    file = f'../build/test/results_{frame}.csv'
    label = '3rd order - Van Leer'
    plt.cla()
    data = np.loadtxt(file, delimiter=',', skiprows=2)
    x = data.transpose()[0][:-2]
    y = data.transpose()[1][:-2]
    line = plt.plot(x, y, label=label)
    ax.legend(title='$t=0.01$ s')
    ax.set_xlabel('x (m)')
    ax.set_ylabel('$\\rho$ (kg/m$^3$)')
    return line


ani = animation.FuncAnimation(fig=fig, func=update, frames=100, interval=100, blit=True)
ani.save('rho.mp4')
```
